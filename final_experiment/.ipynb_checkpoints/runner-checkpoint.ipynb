```python
# Dependancies

import pandas as pd

from experiment_runner_utils import (parse_rna,
                                     parse_metadata,
                                     deseq2_analysis_class,
                                     calculate_and_plot_genesets,
                                     parse_gsea_df,
                                     dna_mutation_class)
```

```python
# Load in RNA data
gene_counts = pd.read_csv(r"C:\Users\judeh\Desktop\lung_cancer\lung_cancer\data\raw_data\data_mrna_seq_v2_rsem.txt", sep='\t')
gene_counts = parse_rna(gene_counts)
gene_counts.head(5)
```

```python
dna_mutations = pd.read_csv(r"C:\Users\judeh\Desktop\lung_cancer\lung_cancer\data\raw_data\TCGA-LUAD.mutect2_snv.tsv", sep='\t')
dna_mutations.head(5)
```

```python
# experimental setup

metadata_filepath = r"C:\Users\judeh\Desktop\lung_cancer\lung_cancer\data\meta_data\dfs_late_stage_meta.csv"
metadata_label_string = 'Disease_Free_Status'
metadata_label_group1 = '0:DiseaseFree'
metadata_label_group2 = '1:Recurred/Progressed'

metadata = parse_metadata(metadata_filepath, metadata_label_string)

analysis = deseq2_analysis_class(gene_counts=gene_counts, metadata=metadata)

# Run DESeq2 analysis
results = analysis.perform_deseq2(
    metadata_column=metadata_label_string,
    group_1=metadata_label_group1,
    group_2=metadata_label_group2
)

volcano_fig = analysis.volcano_plot()
```

```python
gsea_df = parse_gsea_df(results)
```

```python
gene_sets = ['GO_Biological_Process_2025', 'MSigDB_Oncogenic_Signatures', 'GO_Molecular_Function_2025']

for gene_set in gene_sets:

    gsea_results_df, gsea_fig = calculate_and_plot_genesets(
        raw_df=gsea_df,
        n_top_genes=1000,
        plot_n_genesets=10,
        gene_set = gene_set
    )
```
