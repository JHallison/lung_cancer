## Perform machine learning

```python
gene_counts.head(5)
metadata.head(5)
merged_df = metadata.merge(gene_counts, left_index=True, right_index=True)
merged_df.head(5)

#machine learning setup

X = merged_df.drop(columns=[metadata_label_string])
X.columns = X.columns.astype(str)

y = merged_df[metadata_label_string]
y = y.map({metadata_label_group1: 0, metadata_label_group2: 1})

X.head(5), y.head(5)
```

```python
pipeline = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000, random_state=42)
)

# Get out-of-fold predicted probabilities
oof_preds = cross_val_predict(
    pipeline, X, y,
    cv=5,
    method='predict_proba'
)

# Get probabilities for the positive class (usually class 1)
oof_preds_pos = oof_preds[:, 1]
oof_auc = roc_auc_score(y, oof_preds_pos)
print(f"OOF ROC AUC Score: {oof_auc:.4f}")

roc_fig = plot_roc_curve(y, oof_preds_pos)
```

```python
# Save figure

full_path = os.path.join(save_path, "roc_curve_dfs_late_stage.pdf")
roc_fig.savefig(full_path, dpi=300, bbox_inches='tight')
```

## model with DNA

```python
# mutation_counts = dna_mutations.groupby("Sample_ID").size().reset_index(name="total_mutation_count")

effect_counts = dna_mutations.pivot_table(index="Sample_ID",
                                columns="effect",
                                aggfunc="size",
                                fill_value=0).reset_index()


features = mutation_counts.merge(effect_counts, on="Sample_ID")

for col in effect_counts.columns:
    if col != "Sample_ID":
        features[f"{col}_ratio"] = features[col] / features["total_mutation_count"]


gene_counts = dna_mutations.groupby("Sample_ID")["gene"].nunique().reset_index(name="unique_gene_count")
features = features.merge(gene_counts, on="Sample_ID")
features["mutation_per_gene"] = features["total_mutation_count"] / features["unique_gene_count"]

# Final engineered features
print(features.head())
```

```python
merged_dna_df = metadata.merge(features, left_index=True, right_on="Sample_ID", how="left")
merged_dna_df = merged_dna_df.drop(columns=["Sample_ID"])

#drop na on the merged dataframe
merged_dna_df = merged_dna_df.dropna()

X_dna = merged_dna_df.drop(columns=[metadata_label_string])
y_dna = merged_dna_df[metadata_label_string]

X_dna
```

```python
X_dna.shape
```

```python
# Force to be 0 and 1
 
le = LabelEncoder()

y_dna = le.fit_transform(y_dna)
```

```python
# from machine_learning_utils import plot_roc_curve

pipeline = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000, random_state=42)
)

# Get out-of-fold predicted probabilities
oof_preds = cross_val_predict(
    pipeline, X_dna, y_dna,
    cv=5,
    method='predict_proba'
)

# Get probabilities for the positive class (usually class 1)
oof_preds_pos = oof_preds[:, 1]
oof_auc = roc_auc_score(y_dna, oof_preds_pos)
print(f"OOF ROC AUC Score: {oof_auc:.4f}")

dna_roc_curve = plot_roc_curve(y_dna, oof_preds_pos)
```

```python
# Save plot
full_path = os.path.join(save_path, "dna_roc_curve_dfs_early_stage.pdf")
dna_roc_curve.savefig(full_path, dpi=300, bbox_inches='tight')
```

```python
#tune it
study = run_optuna(X_dna, y_dna, n_trials=20)
best_params = study.best_params
print("Best hyperparameters:", best_params)

pipeline = build_lr_pipeline_from_best_params(best_params)

oof_preds = cross_val_predict(
    pipeline, X_dna, y_dna,
    cv=5,
    method='predict_proba'
)

oof_preds_pos = oof_preds[:, 1]
oof_auc = roc_auc_score(y_dna, oof_preds_pos)
print(f"OOF ROC AUC Score: {oof_auc:.4f}")

dna_roc_curve_optuna = plot_roc_curve(y_dna, oof_preds_pos)
```
